```python
grade_list = list(enumerate(df['sub_grade'].sort_values(ascending=False).unique()))
```

```python
def enum(grade):
    for i in grade_list:
        if grade==i[1]:
            return i[0]
```

```python
df['sub_grade_num'] = df['sub_grade'].apply(enum)
```

```python
#X = df[['loan_amnt','int_rate','installment','open_acc',
#        'pub_rec','total_acc','pub_rec_bankruptcies',
#        'grade_num','sub_grade_num']]
X = df[['int_rate','grade_num','sub_grade_num']]

y = df['loan_repaid']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(3,activation='relu'))

model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=500,callbacks=[early_stop],validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```
